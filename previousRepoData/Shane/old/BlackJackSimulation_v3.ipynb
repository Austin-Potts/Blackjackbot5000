### Function: Create a deck

```python
def createPlayingDeckStack(deckOfCards, numberOfDecks):
    # Empty list holding cards
    newDeck = []
    
    # Loop through number of decks
    for i in range(numberOfDecks):
        
        # Extend newDeck List with a new deck of cards, depending on how many decks are te be used
        newDeck.extend(deckOfCards)
        
    # Shuffle the newly assembled deck of cards
#     random.shuffle(newDeck)
        
    # Shuffle the newly assembled deck of cards 7 times    
    for j in range(7):
        rd.shuffle(newDeck)

    # Return the newly shuffled stack of playing cards to be used
    return newDeck
```

### Function: Possible values of the aces within a hand
##### no aces = 0; one ace = 1 or 11; two aces = 2 or 12; etc...

```python
def possibleAceValues(numberOfAces):
    
    # if there are no aces, return a list of one value: 0
    if numberOfAces == 0:
        return [0]
    
    # return a list of two possible values for the aces in the hand:
    # 1st value  = sum of ace(s) if all are to be used as a value of 1;
    # 2nd value = sum of ace(s) where first is value of 11 and remainder(s), if applicable, are value of 1
    else:
        return [numberOfAces, 10 + numberOfAces]
```

### Function: Add up the value of the hand

```python
def addCardValues(handOfCards):
    # Set number of Aces to zero (loop through below to increase in increments)
    numberOfAces = 0
    
#     # Set number of 10 value cards to zero (loop through below to increase in increments)
#     numberOfTens = 0
    
    # Set handSubTotal to zero (loop through below to increase in by card value, when the card isn't an ace)
    handSubTotal = 0
    
    # Loop through each card held within the hand of cards
    for card in handOfCards:
        
        # If card is an ace, then add one to running number of aces within the hand (starting at 0)
        if card == 'A':
            numberOfAces += 1
            
        # If card is a 10, J, Q or K, then add 10 to the running handSubTotal
        elif card in [10, 'J', 'Q', 'K']:
            handSubTotal += 10
            
        # Otherwise, add face value to the running handSubTotal
        else:
            handSubTotal += card
            
    # Create a list of possible values of the ace(s) in the hand by calling possibleAceValues function
    possibleAceValueList = possibleAceValues(numberOfAces)
    
    # Create a list of possible hand total values, using possibleAceValues (at a min, this list is [0])
    handFinalTotal = [i + handSubTotal for i in possibleAceValueList if i + handSubTotal <= 21]
    
    # Evaluate the number of entries within handFinalTotal (possible values) to get hand total
    # If not blank (because total was <= 21), then return max possible value
    if handFinalTotal != []:
        return max(handFinalTotal)
    # Otherwise (if it was blank because total > 21), return the min value of possibleAceValueList + handSubTotal
    else:
        return min(possibleAceValueList) + handSubTotal
        
```

### Function: Coin Toss: Heads or Tails
###### (player decides to hit or stand using this function; Head = "Hit Me!", Tails = "Stand.")

```python
def coinToss():
    coinTossValue = rd.random()
    if coinTossValue >= 0.5:
        return "Heads"
    else:
        return "Tails"
    
```

### Define the number of stacks, decks, players and other variables to be used

```python
numberOfStacks = 10000
# numberOfStacks = 1
numberOfDecks = 6
# numberOfDecks = 1
moveToNewStackThreshold = 52
# moveToNewStackThreshold = 20

numberOfPlayers = 1


# tableMinBet = 5
# tableMaxBet = 500
```

### Create empty lists to hold the dealer's hand, player's hand and player's score/results*********

```python
"""GAME DATA"""
gameStackPlayed = []

cardsInStackAtStart = []

cardsInStackAtEnd = []

cardsPlayed = []

gameBetsTotal = []





"""DEALER DATA"""
# dealerUpCards list holds the dealer's upcard (1st card) dealt for each game, over the course of all games (going through all stacks of cards)
dealerUpCards = []

# dealerHands list holds the dealer's cards dealt for each game, over the course of all games (going through all stacks of cards)
dealerHands = []

# dealerCardTotals list holds the dealer's scores for each game, over the course of all games (going through all stacks of cards)
dealerCardTotals = []

# dealerNumberOfCards list holds the number of cards dealt for each game, over the course of all games (going through all stacks of cards)
dealerNumberOfCards = []

"""PLAYER DATA"""
# playerHands list holds the player's cards dealt for each game, over the course of all games (going through all stacks of cards)
playerHands = []

# playerCardTotals list holds the players scores for each game, over the course of all games (going through all stacks of cards)
playerCardTotals = []

# playerNumberOfCards list holds the number of cards dealt for each game, over the course of all games (going through all stacks of cards)
playerNumberOfCards = []

# playerResults list holds the results of each game, over the course of all games (going through all stacks of cards)
# -1.0 = Player Loss; 0.0 = Push/Tie; 1.0 = Player Win
playerResults = []

playerMoneyBalances = []

playerBets = []

playerWinLossRunningTotals = []

# playerSplit = []

# playerDouble = []


gameCounter = 0
playingDeckStackCounter = 0

moneyToStart = 10000
playerMoney = [moneyToStart for player in range(numberOfPlayers)]
playerNetWinLoss = [0 for player in range(numberOfPlayers)]

playerBaseBet = 5
playerMidBet = 25
playerMaxBet = 100

# playerTotalAmountBet = 0

# playerNetTotal = 0

# Loop through each stack of cards (numberOfStacks; each stack is x decks of cards, or x * numberOfDecks)
for stack in range(numberOfStacks):
    
    # Create cards in current stack (by using createPlayingDeckStack function)
    playingDeckStack = createPlayingDeckStack(deckOfCards, numberOfDecks)
    
    # Increase playingDeckStackCounter by 1
    playingDeckStackCounter += 1
    
    # While loop: loop through while the number of cards in the current stack remains over the threshold amount
    while len(playingDeckStack) > moveToNewStackThreshold:
        
        # Start a new game...
        # Increase gameCounter by 1
        gameCounter += 1
        
        # Setup parameters for each game...
        
        # Set current round/game result for each player to 0 (Push/Tie); acting as a placeholder, updated later
        playerResult = np.zeros((1,numberOfPlayers))
        
        # Set current round/game card total for each player ********to 0********; acting as a placeholder, updated later
        playerCardTotal = [[] for player in range(numberOfPlayers)]
        
        playerHandNumberOfCards = [[] for player in range(numberOfPlayers)]
        
        playerBet = [playerBaseBet for player in range(numberOfPlayers)]
        playerRunningBalance = [0 for player in range(numberOfPlayers)]
        playerRunningNetWinLoss = [0 for player in range(numberOfPlayers)]
        
        # Setup lists that holds the hands dealt for the dealer and each player
        playerHand = [[] for player in range(numberOfPlayers)]
        dealerHand = []
        
        """Player places bet"""
        for player in range(numberOfPlayers):
            playerBet[player] = playerBaseBet
            playerMoney[player] -= playerBet[player]
            playerRunningBalance[player] = playerMoney[player]
            playerNetWinLoss[player] -= playerBet[player]
            playerRunningNetWinLoss[player] = playerNetWinLoss[player]

        # First card for each player...
        for player, hand in enumerate(playerHand):
            playerHand[player].append(playingDeckStack.pop(0))
            
        # First (up) card for the dealer...
        dealerHand.append(playingDeckStack.pop(0))
        
        # Second card for each player...
        for player, hand in enumerate(playerHand):
            playerHand[player].append(playingDeckStack.pop(0))
            
        # Second card for the dealer...
        dealerHand.append(playingDeckStack.pop(0))
        
        # Initial check to see if dealer has blackjack (21) after two cards dealt to each player and dealer...
#         if set(dealerHand) == blackjack:
        if addCardValues(dealerHand) == 21:
            # Loop through each player's initial two cards dealt...
            for player in range(numberOfPlayers):
                # Check to see if each player also has blackjack (21)
#                 if set(playerHand[player]) == blackjack:
                if addCardValues(playerHand[player]) == 21:
                    # If so, game is a Push/Tie, assign result value of 0
                    playerResult[0, player] = 0
                    playerCardTotal[player] = addCardValues(playerHand[player])
                    playerHandNumberOfCards[player] = len(playerHand[player])
                    
                    """Player receives their bet back"""
                    playerMoney[player] += playerBet[player]
                    playerRunningBalance[player] = playerMoney[player]
                    playerNetWinLoss[player] += playerBet[player]
                    playerRunningNetWinLoss[player] = playerNetWinLoss[player]
                    
                # If not, game is a Loss for the player, assign result value of -1
                else:
                    playerResult[0, player] = -1
                    playerCardTotal[player] = addCardValues(playerHand[player])
                    playerHandNumberOfCards[player] = len(playerHand[player])
        
        # If dealer does not have blackjack (21) after initial two cards dealt...
        else:
            # Loop through each player's initial two cards dealt...
            for player in range(numberOfPlayers):
                # Check to see if each player has blackjack (21)
#                 if set(playerHand[player]) == blackjack:
                if addCardValues(playerHand[player]) == 21:
                    # If so, game is a win for that player, assign result value of 1
                    playerResult[0, player] = 1
                    playerCardTotal[player] = addCardValues(playerHand[player])
                    playerHandNumberOfCards[player] = len(playerHand[player])
                    
                    """Player receives 3:2 of their bet in winnings"""
                    playerMoney[player] += 1.5*playerBet[player] + playerBet[player]
                    playerRunningBalance[player] = playerMoney[player]
                    playerNetWinLoss[player] += 1.5*playerBet[player] + playerBet[player]
                    playerRunningNetWinLoss[player] = playerNetWinLoss[player]
                else:
                    
                    
                    # Double Down? ******************************
#                     TBD
                    
                    # If not, player will randomly decide to hit or stand using coin toss function...
                    while addCardValues(playerHand[player]) < 21 and coinToss() == "Heads":
                        # Deal another card to the player...
                        playerHand[player].append(playingDeckStack.pop(0))
                        
                        
                # Dealer to hit or stand depending on house score so long as all players haven't busted
                while addCardValues(playerHand[player]) <= 21 and addCardValues(dealerHand) < 17:
                    dealerHand.append(playingDeckStack.pop(0))

                if addCardValues(playerHand[player]) > 21:
                    playerResult[0, player] = -1
                    playerCardTotal[player] = addCardValues(playerHand[player])
                    playerHandNumberOfCards[player] = len(playerHand[player])
                elif addCardValues(dealerHand) > 21 or\
                        addCardValues(playerHand[player]) > addCardValues(dealerHand):
                    playerResult[0, player] = 1
                    playerCardTotal[player] = addCardValues(playerHand[player])
                    playerHandNumberOfCards[player] = len(playerHand[player])
                    """Player receives 1:1 of their bet in winnings"""
                    playerMoney[player] += 2*playerBet[player]
                    playerRunningBalance[player] = playerMoney[player]
                    playerNetWinLoss[player] += 2*playerBet[player]
                    playerRunningNetWinLoss[player] = playerNetWinLoss[player]
                elif addCardValues(playerHand[player]) == addCardValues(dealerHand):
                    playerResult[0, player] = 0
                    playerCardTotal[player] = addCardValues(playerHand[player])
                    playerHandNumberOfCards[player] = len(playerHand[player])
                    """Player receives their bet back"""
                    playerMoney[player] += playerBet[player]
                    playerRunningBalance[player] = playerMoney[player]
                    playerNetWinLoss[player] += playerBet[player]
                    playerRunningNetWinLoss[player] = playerNetWinLoss[player]
                else:
                    playerResult[0, player] = -1
                    playerCardTotal[player] = addCardValues(playerHand[player])
                    playerHandNumberOfCards[player] = len(playerHand[player])
                    
        # Append list of hands
        
        dealerUpCards.append(dealerHand[0])
        dealerHands.append(dealerHand)
        dealerCardTotals.append(addCardValues(dealerHand))
        dealerNumberOfCards.append(len(dealerHand))
        playerHands.append(playerHand)
        playerCardTotals.append(playerCardTotal)
        playerNumberOfCards.append(playerHandNumberOfCards)
        playerResults.append(list(playerResult[0]))

        playerBets.append(playerBet)
        playerMoneyBalances.append(playerRunningBalance)
        playerWinLossRunningTotals.append(playerRunningNetWinLoss)
                
            
        
        
        
            
            
```
